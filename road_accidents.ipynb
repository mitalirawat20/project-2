```python
#correlation coefficient of two data
def calculate_correlation(data, column1, column2):
    correlation_coefficient= data[column1].corr(data[column2])
    return correlation_coefficient

column1="Fine - Total Acc. - 2014"
column2="Fine - Persons Killed - 2014"
correlation = calculate_correlation(data, column1, column2)
print(f'The correlation coefficient between {column1} and {column2} is: {correlation}')
```

```python
#normalisation distribution
def normal_dist(x,mean,sd):
    prob_density=(np.pi*sd)*np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density

mean =0
sd=1
x=1
result=normal_dist(x,mean,sd)
print(result)
```

```python
#baye's theorem
def bayes_theorem(p_A_given_B,  p_B_given_A, p_A):

  numerator= p_A_given_B * p_A
  denominator= p_B_given_A * p_A+ (1- p_B_given_A)*(1-p_A)

  if denominator !=0:
    return numerator/denominator
  else:
    return 0

#example usage
p_A_given_B=0.8
p_B_given_A=0.6
p_A=0.3

updated_probablity= bayes_theorem(p_A_given_B,  p_B_given_A, p_A)
print("updated probablity", updated_probablity)
```
